```python
# import
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
#W is the wiener varibale
# W[0]=0
# W[t]=w[t-1]*sqrt(dt)*N(0,1)
T=1 # start from position 1
N=500 # simulate 500 values
dt=T/N # delta
dW = np.zeros(N) # make an empty array    
W = np.zeros(N)  # make an empty array   
dW[1] =math.sqrt(dt)*np.random.randn()  # calculate second  element (the firts element is 0)
W[1] = dW[1]              
for j in range(2,N):
    dW[j] = math.sqrt(dt)*np.random.randn()
    W[j] = W[j-1] + dW[j]   # generate the elements
```

```python
plt.figure(figsize=(12,8))
plt.plot(W,'r-')   
plt.xlabel('t') 
plt.ylabel('W(t)')
```

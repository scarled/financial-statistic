```python
# import
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
N=5000# simulate 5000 values   
W = np.zeros(N)  # make an empty array   
W[1] = np.random.choice([-1,1])
d = np.zeros(N)  # make an empty array   
d[1]=W[1]
for j in range(2,N):
    d[j]= np.random.choice([-1,1])
    W[j] = W[j-1] +d[j]

```

```python
plt.figure(figsize=(12,8))
plt.plot(W,'r-')   
plt.xlabel('t') 
plt.ylabel('W(t)')
np.mean(W)
```
